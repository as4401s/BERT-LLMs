```python
from transformers import BertForNextSentencePrediction, BertTokenizer
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_nsp = BertForNextSentencePrediction.from_pretrained('bert-base-uncased')
```

```python
bert_nsp
```

```python
text1 = 'I had visited Turkey this week.'
text2 = 'I specifically liked Cappadocia, Istanbul and the food there'
```

```python
inputs = tokenizer(text1, text2, return_tensors='pt')
```

```python
inputs
```

```python
inputs.input_ids
```

```python
outputs = bert_nsp(**inputs)
outputs
```

```python
outputs = bert_nsp(**inputs, labels=torch.LongTensor([0])) #0 - true, 1-false
outputs
```

```python
outputs = bert_nsp(**inputs, labels=torch.LongTensor([1])) #0 - true, 1-false
outputs
```
