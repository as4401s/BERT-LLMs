```python
from transformers import pipeline, BertForQuestionAnswering
from transformers import BertForTokenClassification, BertForSequenceClassification
```

## Question Answering

```python
bert_qa = BertForQuestionAnswering.from_pretrained('bert-base-uncased')
bert_qa
```

```python
bert_qa.qa_outputs
```

```python
model_name = 'deepset/roberta-base-squad2'
qa = pipeline(model=model_name, tokenizer=model_name, revision='v1.0', task='question-answering')
```

```python
sequence = 'Where is Arjun living these days?','Arjun lives in Jena but Ayantika lives in Berlin'
```

```python
qa(*sequence)
```
